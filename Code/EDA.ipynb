## Multivariate

```python
# Create decade column
data['decade'] = (data['release_date'] // 10) * 10
```

```python
# Filter for top 6 artists
top_artists = data['artist_name'].value_counts().head(6).index
data_top = data[data['artist_name'].isin(top_artists)]


# Group data
grouped = data_top.groupby(['artist_name', 'decade', 'topic']).size().reset_index(name='count')


# Plot Facet Grid
g = sns.FacetGrid(grouped, col='artist_name', col_wrap=2, height=5, sharex=False, sharey=False)
g.map_dataframe(sns.barplot, x='decade', y='count', hue='topic', palette=custom_colors)
g.set_titles("{col_name}")
g.add_legend(title="Topic")

for ax in g.axes.flat:
    ax.set_xlabel("Decade")
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Correlation of all numeric features with "decade"
corr_decade= data.corr(numeric_only=True)['decade'].sort_values(ascending=False)
print(corr_decade)

# Drop "Unnamed:0" and "decade" column if present
num_features= data.drop(data.filter(regex="Unnamed").columns, axis=1)

if 'decade' in data.columns:
    data = data.drop(columns=['decade'])
# Select only Numerical Features
num_features= data.select_dtypes(include=[np.number])

# Plot heatmap
plt.figure(figsize=(17,14))
sns.heatmap(num_features.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## Summary & Important Takeaways

##### At first, I only planned to drop the 'decade' column, but I'll more than likely drop the 'release_date' column too. Both columns are very similar to 'age' as they're highly correlated with a high negative correlation. For example, as the release year goes up, a song's age goes down. So keeping all three would be repetitive, and which is why I'm only keeping 'age' to simplify and avoid any confusion during my modeling

##### 'len' shows a moderately positive correlation with 'obscene', meaning longer songs  may include more explicit content. 'len' also has a fairly weak negative correlation with 'age'. suggesting newer songs tend to be slightly longer

##### Violence and sadness show a weak negative correlation, meaning that songs with violence focus less on sadness. This aligns with my "Music Trends by Topic" charts, where we see a clear shift in popular topics over team and also a clear seperation between the more emotional sad songs and violence
